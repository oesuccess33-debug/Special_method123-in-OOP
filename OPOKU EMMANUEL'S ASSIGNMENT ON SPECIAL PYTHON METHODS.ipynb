# Python Special Methods

Dunder = **D**ouble **UNDER**score. Like `__init__` or `__str__`.

These methods let your classes work with Python's built-in stuff like `print()`, `len()`, `==`, and loops.

## `__init__` — Set Up Your Object

Runs automatically when you create an object. Use it to set starting values.

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

buddy = Dog("Buddy", 3)
print(f"{buddy.name} is {buddy.age} years old")
```

## `__str__` and `__repr__` — String Output

- `__str__`: What you see when you `print()` the object
- `__repr__`: What developers see (should show how to recreate the object)

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
    
    def __str__(self):
        return f"{self.title} by {self.author}"
    
    def __repr__(self):
        return f"Book('{self.title}', '{self.author}')"

b = Book("1984", "George Orwell")
print(str(b))   # For users
print(repr(b))  # For devs
```

## `__len__` — Make `len()` Work

Lets you use `len()` on your object. Must return an integer.

```python
class Backpack:
    def __init__(self):
        self.items = []
    
    def add(self, item):
        self.items.append(item)
    
    def __len__(self):
        return len(self.items)

pack = Backpack()
pack.add("water")
pack.add("snacks")
pack.add("map")
print(f"Items: {len(pack)}")
```

## `__eq__` — Compare with `==`

By default, `==` checks if two variables point to the same object.
With `__eq__`, you can compare by content instead.

```python
class Person:
    def __init__(self, id, name):
        self.id = id
        self.name = name
    
    def __eq__(self, other):
        return self.id == other.id

p1 = Person(1, "Alice")
p2 = Person(1, "Alice")
print(f"Same content: {p1 == p2}")   # True
print(f"Same object: {p1 is p2}")    # False
```

## `__add__`, `__sub__`, `__mul__` — Math Operators

Teach Python what `+`, `-`, `*` mean for your objects.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)
    
    def __str__(self):
        return f"({self.x}, {self.y})"

a = Point(2, 3)
b = Point(4, 5)
print(f"{a} + {b} = {a + b}")
```

## `__getitem__` — Index with `[]`

Lets you use bracket notation: `obj[0]`, `obj[1]`, etc.

```python
class Playlist:
    def __init__(self):
        self.songs = []
    
    def add(self, song):
        self.songs.append(song)
    
    def __getitem__(self, index):
        return self.songs[index]

pl = Playlist()
pl.add("Song A")
pl.add("Song B")
pl.add("Song C")
print(pl[0])
print(pl[-1])
```

## `__setitem__` — Assign with `[]`

Lets you do `obj[0] = value`.

```python
class Playlist:
    def __init__(self):
        self.songs = []
    
    def add(self, song):
        self.songs.append(song)
    
    def __getitem__(self, index):
        return self.songs[index]
    
    def __setitem__(self, index, value):
        self.songs[index] = value

pl = Playlist()
pl.add("Old Song")
pl.add("Another")
print(f"Before: {pl.songs}")
pl[0] = "New Song"
print(f"After:  {pl.songs}")
```

## `__contains__` — The `in` Operator

Defines what happens when you use `in` on your object.

```python
class Fridge:
    def __init__(self):
        self.items = []
    
    def put(self, item):
        self.items.append(item)
    
    def __contains__(self, item):
        return item in self.items

fridge = Fridge()
fridge.put("milk")
fridge.put("eggs")
print(f"Got milk? {'milk' in fridge}")
print(f"Got pizza? {'pizza' in fridge}")
```

## `__call__` — Call Object Like a Function

Use `obj()` to call your object. Good for things that need settings but act like functions.

```python
class Multiplier:
    def __init__(self, factor):
        self.factor = factor
    
    def __call__(self, number):
        return number * self.factor

double = Multiplier(2)
triple = Multiplier(3)
print(f"Double 5: {double(5)}")
print(f"Triple 5: {triple(5)}")
```

## `__iter__` and `__next__` — Loop with `for`

Makes your object work in `for` loops.

- `__iter__` returns the iterator (usually `self`)
- `__next__` returns the next value or raises `StopIteration` when done

```python
class Countdown:
    def __init__(self, start):
        self.start = start
    
    def __iter__(self):
        self.current = self.start
        return self
    
    def __next__(self):
        if self.current < 0:
            raise StopIteration
        num = self.current
        self.current -= 1
        return num

for n in Countdown(5):
    print(n, end=" ")
print("Blast off!")
```

## `__bool__` — True or False?

Controls whether your object is truthy or falsy in `if` statements.

```python
class Wallet:
    def __init__(self, cash):
        self.cash = cash
    
    def __bool__(self):
        return self.cash > 0

w1 = Wallet(50)
w2 = Wallet(0)
print(f"Has money: {bool(w1)}")
print(f"Has money: {bool(w2)}")
```

## `__del__` — Cleanup (Optional)

Called when an object is about to be destroyed.

**Note:** Don't rely on this for important cleanup. Use `with` statements instead.

```python
class TempFile:
    def __init__(self, name):
        self.name = name
        print(f"Created: {name}")
    
    def __del__(self):
        print(f"Cleaned up: {self.name}")

f = TempFile("data.txt")
del f
```

| `__del__` | Cleanup | `del obj` | `__del__` |

# Remember

1. You don't call these methods directly — Python does it for you.
2. `__init__` sets up your object when it's created.
3. `__str__` is for users; `__repr__` is for debugging.
4. `__eq__` compares content, not memory address.
5. Operator methods (`__add__`, etc.) make your types feel natural.
6. `__getitem__`/`__setitem__` let you use `[]` on your objects.
7. `__iter__`/`__next__` let your objects work in `for` loops.
8. `__call__` makes objects act like functions.
9. `__bool__` controls truthiness.
10. Don't rely on `__del__` for cleanup — use `with` instead.

**The big idea:** If Python has a built-in way to interact with data, there's a dunder method that lets your class join in.
